```python
import pandas as pd
import numpy as np
import random
import uuid

# Number of synthetic customers
N = 100

# Churn dataset columns
CreditScore = np.random.randint(350, 850, N)
Geography = np.random.choice(['France', 'Germany', 'Spain'], N)
Gender = np.random.choice(['Male', 'Female'], N)
Age = np.random.randint(18, 90, N)
Tenure = np.random.randint(0, 11, N)
Balance = np.round(np.random.uniform(0, 250000, N), 2)
NumOfProducts = np.random.randint(1, 5, N)
HasCrCard = np.random.choice([0,1], N)
IsActiveMember = np.random.choice([0,1], N)
EstimatedSalary = np.round(np.random.uniform(20000, 150000, N), 2)

# Loan Default dataset columns
Income = np.random.randint(20000, 150000, N)
LoanAmount = np.random.randint(5000, 50000, N)
MonthsEmployed = np.random.randint(1, 360, N)
NumCreditLines = np.random.randint(1, 10, N)
InterestRate = np.round(np.random.uniform(3.0, 12.0, N),2)
LoanTerm = np.random.choice([12, 24, 36, 48, 60], N)
DTIRatio = np.round(np.random.uniform(0.1, 0.5, N),2)
Education = np.random.choice(['High School', 'Bachelor', 'Master', 'PhD'], N)
EmploymentType = np.random.choice(['Salaried', 'Self-Employed', 'Unemployed', 'Retired'], N)
MaritalStatus = np.random.choice(['Single', 'Married', 'Divorced'], N)
HasMortgage = np.random.choice(['Yes','No'], N)
HasDependents = np.random.choice(['Yes','No'], N)
LoanPurpose = np.random.choice(['Home','Car','Business','Education','Other'], N)
HasCoSigner = np.random.choice(['Yes','No'], N)

# -----------------------
# Create combined DataFrame
# -----------------------
new_customers = pd.DataFrame({
    'CreditScore': CreditScore,
    'Age': Age,
    'Geography': Geography,
    'Gender': Gender,
    'Tenure': Tenure,
    'Balance': Balance,
    'NumOfProducts': NumOfProducts,
    'HasCrCard': HasCrCard,
    'IsActiveMember': IsActiveMember,
    'EstimatedSalary': EstimatedSalary,
    'Income': Income,
    'LoanAmount': LoanAmount,
    'MonthsEmployed': MonthsEmployed,
    'NumCreditLines': NumCreditLines,
    'InterestRate': InterestRate,
    'LoanTerm': LoanTerm,
    'DTIRatio': DTIRatio,
    'Education': Education,
    'EmploymentType': EmploymentType,
    'MaritalStatus': MaritalStatus,
    'HasMortgage': HasMortgage,
    'HasDependents': HasDependents,
    'LoanPurpose': LoanPurpose,
    'HasCoSigner': HasCoSigner,
})

# Save to CSV
new_customers.to_csv("C:/AMOD/Extensive_bank/data/combined_customer_data.csv", index=False)
print("Synthetic customer data saved to 'data/combined_customer_data.csv'")
new_customers.head()
```
